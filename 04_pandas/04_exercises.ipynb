#### Exercise 7

1. Use the `map` method to create a new column named `is_expensive` that is 1 if the order cost is greater than 30 and is 0 otherwise.
2. Use the `apply` method to create a new column named `fast_favorite` which is 1 if the cuisine type is `"Japanese"` and the cost is below 20 dollars and is 0 otherwise


```python
df["is_expensive"] = df["order_cost"].map(lambda x: 1 if x > 30 else 0)
df["fast_favorite"] = df[["cuisine_type", "order_cost"]].apply(lambda x: 1 if x["cuisine_type"] == "Japanese" and x["order_cost"] < 20 else 0, axis=1)
```

#### Exercise 11

Using the Pandas DataFrames provided to you, complete the following:

1. Merge `df_orders` and `df_customers` on `customer_id` using an inner join.
2. How many rows are in the resulting DataFrame?
3. Perform a left join using `df_orders` as the left DataFrame. What differences do you observe compared to the inner join?

```python
df_orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4],
    'customer_id': [101, 102, 103, 104],
    'cost': [25, 40, 15, 30]
})

df_customers = pd.DataFrame({
    'customer_id': [101, 102, 103, 105],
    'customer_name': ['Alice', 'Bob', 'Charlie', 'Diana']
})
```

```python
df_inner = pd.merge(df_orders, df_customers, on="customer_id", how="inner")
print(len(df_inner))
df_left = pd.merge(df_orders, df_customers, on="customer_id", how="left")
print(len(df_left))
# 3 rows were in the inner join while there are 4 in the left join
```
