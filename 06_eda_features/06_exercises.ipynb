#### Exercise 7

Make two visualizations that let you to explore the relationships between the following variables at the same time:

- `Salary`
- `EngagementSurvey`
- `EmpSatisfaction`
- `Absences`

```python
sns.pairplot(data=df, vars=["Salary", "EngagementSurvey", "EmpSatisfaction", "Absences"])
```

```python
corr_matrix = df[["Salary", "EngagementSurvey", "EmpSatisfaction", "Absences"]].corr()

sns.heatmap(corr_matrix, annot=True)
```
