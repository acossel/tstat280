### Quiz

#### Part 1

Load the dataset found in `hr.csv`. Then create a visualization that would easily allow you to answer the following question:

- Which performance score is most common?

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("..\data\hr.csv")
sns.histplot(data = df, x = "PerformanceScore")
```

#### Part 2

Create a visualization that would allow you to easily answer the following question:

- What does the distribution of the `Salary` of an employee look like?

```python
sns.boxplot(data = df, x = "Salary")
```

#### Part 3

Create a visualization that would allow you to easily answer the following question:

- How does the distribution of `EngagementSurvey` differ for the categories in `PerformanceScore`?

```python
sns.boxplot(data = df, x = "EngagementSurvey", y = "PerformanceScore") 
```

#### Part 4

Create a visualization that would allow you to easily answer the following question:

- What does the relationship between the log of `Salary` and `Absences` look like?

```python
import numpy as np
df["log_Salary"] = np.log(df["Salary"])
df["log_Absences"] = np.log(df["Absences"])

sns.scatterplot(data = df, x = "log_Salary", y = "log_Absences")
```
