To submit your own work, you need your *own copy* of the course repository.

## What is a Fork?

A **fork** creates a copy of a repository under *your own GitHub account*.

After you fork:

- You can make changes freely
- You can push updates
- You control your version
- The original course repository remains unchanged

---

## Why We Use Forking in This Course

If you clone the instructor's repository directly, you **cannot push changes** to it.

By forking:

- You create your own version
- You push to your version
- You submit the link to your repository

# Virtual Environments

A virtual environment is an isolated Python environment. These are useful because different projects may require:

- Different package versions
- Different dependencies

Virtual environments prevent conflicts between projects.

# Creating a Virtual Environment

A virtual environment keeps your project’s Python packages isolated from other projects.

---

## Step 1: Navigate to Your Project Folder

Open your terminal and move into your project directory:

```bash
cd my_project
```

---

## Step 2: Create the Virtual Environment

Run:

```bash
python -m venv venv
```

This creates a new folder named `venv` inside your project.

---

## Step 3: Activate the Virtual Environment

### 🍎 On Mac / Linux

```bash
source venv/bin/activate
```

### 🪟 On Windows (PowerShell)

```powershell
venv\Scripts\activate
```

---

## ✅ Confirm It Worked

You should now see `(venv)` at the beginning of your terminal line, for example:

```text
(venv) user@computer my_project %
```

This means your virtual environment is active.

---

## Deactivate the Virtual Environment

When you're finished working, run:

```bash
deactivate
```
