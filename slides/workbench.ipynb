# Basic Terminal Commands

Here are the most important commands you’ll use:

| Command | What It Does |
|----------|--------------|
| `ls` | Lists files in the current directory |
| `cd folder_name` | Changes into a directory named `folder_name` |
| `cd ..` | Moves up one directory |
| `mkdir folder_name` | Creates a new directory named `folder_name` |
| `pwd` | Prints the current working directory |
| `clear `| Clears the console |
| `rm` | Removes a file from the directory |

# Forking a Repository

To submit your own work, you need your *own copy* of the course repository.

We do this using something called a **fork**.

<div>
<img src="images/github_fork.png" width="500" alt="Repository page with fork button highlighted"/>
</div>

# What is a Fork?

A **fork** creates a copy of a repository under *your own GitHub account*.

After you fork:

- You can make changes freely
- You can push updates
- You control your version
- The original course repository remains unchanged

# Why Fork?

If you clone the instructor's repository directly, you **cannot push changes** to it.

By forking:

- You create your own version
- You push to your version
- You submit the link to your repository

# Virtual Environments

A virtual environment is an isolated Python environment. These are useful because different projects may require:

- Different package versions
- Different dependencies

Virtual environments prevent conflicts between projects.

# Creating a Virtual Environment

A virtual environment keeps your project’s Python packages isolated from other projects. Open your terminal and move into your project directory:

```bash
cd my_project
```

This is the folder where we want to create the virtual environment.

# Create the Virtual Environment

Run:

```bash
python -m venv venv
```

This creates a new folder named `venv` inside your project.

<div>
<img src="images/terminal_venv.png" width="500" alt="Terminal showing newly created virtual environment"/>
</div>

# Activate the Virtual Environment

### On Mac / Linux

```bash
source venv/bin/activate
```

### On Windows (PowerShell)

```powershell
venv\Scripts\activate
```

<div>
<img src="images/terminal_active_venv.png" width="500" alt="Terminal showing newly activated virtual environment"/>
</div>

## Confirm It Worked

You should now see `(venv)` at the beginning of your terminal line, for example:

```text
(venv) user@computer my_project %
```

This means your virtual environment is active. Don't do this right now, but when you're finished working, in the terminal run:

```bash
deactivate
```

# Install Data Science Libraries

Once your virtual environment is activated, install the required libraries by running:

```bash
pip install numpy pandas matplotlib seaborn scikit-learn jupyterlab
```

This command installs most required packages for the course in one step.
