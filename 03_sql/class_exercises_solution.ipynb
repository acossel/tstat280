#### Exercise 2

Now, do the following:

1. Import the `sqlite3` library
2. Create a connection to the database
3. Create a cursor you can use for querying the database

```python
import sqlite3

conn = sqlite3.connect("murder_mystery.db")
cur = conn.cursor()
```

#### Exercise 3

Select the first 5 records from the `drivers_license` table.

```python
cur.execute("""
    SELECT * FROM drivers_license
    LIMIT 5
""").fetchall()
```

#### Exercise 4

Find the following pieces of information:

1. The average age of individuals in `drivers_license` (name it average_age)
2. The average height of individuals in `drivers_license` (name it average_weight)
3. The minimum annual income in `income` (name it minimum_income)
4. The maximum annual income in `income` (name it maximum_income)


```python
res = cur.execute("""
    SELECT AVG(age) as average_age, AVG(height) as average_height FROM drivers_license
""").fetchall()

print(res)

res = cur.execute("""
    SELECT MIN(annual_income) as minimum_income, MAX(annual_income) as maximum_income FROM income
""").fetchall()

print(res)
```

#### Exercise 5

Select the first 4 individuals from the `drivers_license` table that have an age over 60 and a height under 60 or that have a license plate number that starts with a "P".

```python
cur.execute("""
    SELECT * FROM drivers_license
    WHERE (age > 60 AND height < 60) OR (plate_number LIKE "P%")
    LIMIT 4
""").fetchall()
```

#### Exercise 6

Do the following:

1. Get the average age of all individuals in `drivers_license` broken down by eye color, hair color, and gender.
2. Only return the groups with an average age above 55
3. Order these groups from largest to smallest average height

```python
cur.execute("""
    SELECT eye_color, hair_color, gender, AVG(age) as average_age FROM drivers_license
    GROUP BY eye_color, hair_color, gender
    HAVING average_age > 55
    ORDER BY average_age DESC
""").fetchall()
```

#### Exercise 7

Join the `person` and `income` tables so that, in addition to all details about a person, you also have their income if it is available. Only grab the name, social security number, and income for each person and only show the first 5 results.

```python
cur.execute("""
    SELECT person.name, person.ssn, income.annual_income FROM person
    LEFT JOIN income
    ON person.ssn = income.ssn
    LIMIT 4
""").fetchall()
```

## Solving the Murdery Mystery

We will now proceed to solve the murdery mystery!

A crime has taken place and the detective needs your help. The detective gave you the crime scene report, but you somehow lost it. You vaguely remember that the crime was a ​murder​ that occurred sometime on ​Jan.15, 2018​ and that it took place in ​SQL City​. Start by retrieving the corresponding crime scene report from the police department’s database.

#### Exercise 1

Retrieve the crime scene report. Query the `crime_scene_report` table to retrieve the report. You will want to use the clues regarding the following:

- What time of crime it was
- The date it took place
- Where the crime took place


```python
cur.execute("""
    SELECT description FROM crime_scene_report
    WHERE type == "murder" AND date == "20180115" AND city == "SQL City"
""").fetchall()
```

#### Exercise 2

We will now find the witnesses described in the crime scene report. Use the information given about them to find them in the `person` table.

```python
cur.execute("""
    SELECT * FROM person
    WHERE address_street_name == "Northwestern Dr" OR 
        (address_street_name == "Franklin Ave" AND name LIKE "Annabel%")
    ORDER BY address_street_name ASC, address_number DESC
    LIMIT 2
""").fetchall()
```

#### Exercise 3

We want to find the information that the witnesses provided about the crime. Use the information we have about the witnesses to find their interviews in the `interview` table.

```python
cur.execute("""
    SELECT transcript FROM interview
    WHERE person_id IN (14887, 16371)
""").fetchall()
```

#### Exercise 4

The witnesses provided several important pieces of information about the accused:

- Part of a membership number for a gym
- Part of the license plate number for their car
- The date for another time that they visited the gym

Use the `get_fit_now_member`, `get_fit_now_check_in`, and `drivers_license` tables to get more information about the accused.

```python
cur.execute("""
    SELECT get_fit_now_check_in.check_in_date,
        get_fit_now_member.membership_status,
        get_fit_now_member.id,
        person.id,
        person.name,
        drivers_license.plate_number
    FROM get_fit_now_check_in
    LEFT JOIN get_fit_now_member
    ON get_fit_now_check_in.membership_id = get_fit_now_member.id
    LEFT JOIN person
    ON get_fit_now_member.person_id = person.id
    LEFT JOIN drivers_license
    ON person.license_id = drivers_license.id
    WHERE get_fit_now_check_in.check_in_date == "20180109"
        AND get_fit_now_member.membership_status == "gold"
        AND get_fit_now_member.id LIKE "48Z%"
        AND drivers_license.plate_number LIKE "%H42W%"
""").fetchall()
```

#### Exercise 5

You should now have a name of the individual the witnesses identified at the crime scene. You'll need to go back and look at the interview with this person found in the `interview` table.

```python
cur.execute("""
    SELECT transcript FROM interview
    WHERE person_id = 67318
""").fetchall()
```

#### Exercise 6

From the interview, you've been given a description of the person that hired the individual that committed the crime. You know the following about this individual:

- Their gender
- Their height
- Their hair color
- The car they drive
- An event that they attended along with when and how often they attended the event

Using the information found in the `drivers_license` and `facebook_event_checkin` tables, pinpoint who this individual is.

```python
cur.execute("""
    SELECT facebook_event_checkin.event_name,
    COUNT(facebook_event_checkin.date) AS facebook_event_checkin_num_events,
    person.id,
    person.name,
    drivers_license.height,
    drivers_license.hair_color,
    drivers_license.gender,
    drivers_license.car_make
    FROM facebook_event_checkin
    LEFT JOIN person
    ON facebook_event_checkin.person_id = person.id
    LEFT JOIN drivers_license
    ON person.license_id = drivers_license.id
    WHERE (drivers_license.height >= 65 AND drivers_license.height <= 67)
        AND drivers_license.hair_color = "red"
        AND drivers_license.car_make == "Tesla"
        AND drivers_license.car_model == "Model S"
        AND facebook_event_checkin.event_name == "SQL Symphony Concert"
""").fetchall()
```
